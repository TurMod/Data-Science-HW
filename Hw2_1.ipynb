```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

url = 'https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8'

data = pd.read_html(url, match='Коефіцієнт народжуваності в регіонах України')
```

```python
data[0].head(2)
```

```python
data[0].shape
```

```python
data = data[0].replace('—', np.nan)
data
```

```python
data.dtypes
```

```python
data[['2014', '2019']] = data[['2014', '2019']].apply(pd.to_numeric)
data
```

```python
data.isnull().sum()
```

```python
data = data.drop(27)
data
```

```python
data = data.fillna({'1950': data['1950'].mean(), '1960': data['1960'].mean(), '1970': data['1970'].mean(), '2014': data['2014'].mean(), '2019': data['2019'].mean()})
data
```

```python
data.loc[data['2019'] > data['2019'].mean(), ['Регіон', '2019']]
```

```python
data.loc[data['2014'] == data['2014'].max(), ['Регіон', '2014']]
```

```python
data.plot.bar(x='Регіон', y='2019')
```

```python
import seaborn as sns

sns.stripplot(x='Регіон',y='1950' , data=data)
```

```python
sns.pointplot(x='Регіон',y='1960' , data=data)
```
