```python
import pandas as pd

data = pd.read_csv('2017_jun_final.csv')
```

```python
data.head(5)
```

```python
data.dtypes
```

```python
data.isnull().sum()
```

```python
tmp = data[data.columns.difference(['Язык.программирования'])].isna().any()
data = data.drop(tmp.index[tmp], axis=1)
data
```

```python
data.isnull().sum()
```

```python
data = data.dropna()
data
```

```python
data.shape
```

```python
python_data = data.loc[data['Язык.программирования'] == 'Python']
python_data
```

```python
python_data.shape
```

```python
python_data
```

```python
python_data_agg = python_data.groupby(['Должность'])
```

```python
python_data_agg = python_data.agg({'Зарплата.в.месяц': ['min', 'max']})
python_data_agg
```

```python
def fill_avg_salary(row):
  avg_salary = row['Зарплата.в.месяц'].mean()
  return avg_salary

python_data_agg['avg'] = python_data_agg.apply(fill_avg_salary, axis=1)
python_data_agg
```

```python
python_data_agg.describe()
```

```python
python_data_agg.to_csv('python_data_agg.csv')
```
