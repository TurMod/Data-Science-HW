```python
top_2015 = data[data['year'] == 2015]
top_2015[top_2015['price'] == top_2015['price'].max()]
```

З тих книг, що потрапили до Топ-50 у 2015 році, яка книга найдорожча (можна використати проміжний датафрейм)? **Відповідь: Publication Manual of the American Psychologic 46$**

```python
data[(data['genre'] == 'Fiction') & (data['year'] == 2010)].count()
```

Скільки книг жанру Fiction потрапили до Топ-50 у 2010 році (використовуйте &)? **Відповідь: 20**

```python
data[(data['user_rating'] == 4.9) & ((data['year'] == 2010) | (data['year'] == 2011))]
```

Скільки книг з рейтингом 4.9 потрапило до рейтингу у 2010 та 2011 роках (використовуйте | або функцію isin)? Відповідь: 1

```python
data[(data['year'] == 2015) & (data['price'] < 8)].sort_values('price')
```

Яка книга остання у відсортованому списку? **Відповідь: Old School (Diary of a Wimpy Kid #10)**

```python
df_genre = data.groupby('genre').agg({'price': ['min', 'max']})
df_genre
```

Максимальна ціна для жанру Fiction: Відповідь: 82$

Мінімальна ціна для жанру Fiction: Відповідь: 0$

Максимальна ціна для жанру Non Fiction: Відповідь: 105$

Мінімальна ціна для жанру Non Fiction: Відповідь: 0$

```python
df_authors = data.groupby('author').agg({'name': 'count'})
df_authors
```

Якої розмірності вийшла таблиця? Відповідь: (248, 1)

```python
df_authors[df_authors['name'] == df_authors['name'].max()]
```

Який автор має найбільше книг? Відповідь: Jeff Kinney

Скільки книг цього автора?
Відповідь: 12

```python
df_avg_rating = data.groupby('author').agg({'user_rating': 'mean'})
df_avg_rating
```

```python
df_avg_rating[df_avg_rating['user_rating'] == df_avg_rating['user_rating'].min()]
```

У якого автора середній рейтинг мінімальний? Відповідь: Donna Tartt

Який у цього автора середній рейтинг? Відповідь: 3.9

```python
df_concat = pd.concat([df_authors, df_avg_rating], axis=1)
df_concat.columns = ['amount_of_books', 'user_rating']
df_concat
```

```python
df_concat.sort_values(['amount_of_books', 'user_rating'])
```

Який автор перший у списку?

Відповідь: Muriel Barbery

```python
df_concat.plot()
plt.show()
```
